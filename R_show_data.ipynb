```r
show_data <- function(filename){    
dataset<-read.table(filename,header=TRUE,sep=',')
if(ncol(dataset)>25){
    #直接打印下面一句话
    print(table(dataset[,ncol(dataset)]-1))
}else{
 print(table(dataset$class))   
}
#md.pattern(dataset)缺失值模式
aggr(dataset,prop=FALSE,numbers=TRUE)
matrixplot(dataset)
}
```
